## 2. 사용자가 1부터 9까지의  숫자를 무작위로 입력한후,1부터 9까지의 숫자가 몇개씩 입력되었는지를 확인 할 수 있는 프로그램을 작성하시오.
- (입력)1 1 2 3 4 5 5 5 5 6 6 6 8
- (결과)1:2,2:1,3:1,4:1,5:4,6:3,8:!

```python
data = input('숫자를 공백으로 구분하여 입력하세요:').split() # split() 함수는 리스트를 반환 
print(data)

counter = {}
for ch in data:
    num = int(ch)
    if num in counter.keys():
        counter[num] = counter[num] + 1
    else:
        counter[num] = 1  
        
for k,v in counter.items():
    print('%d : %d'%(k,v))
```

### 3.아래와 같은 문장이 주어져 있을 때, 문장 속에 포함되어 있는단어의 빈도수를 구하는 프로그램을 작성하시오,또 빈도수를 기준으로 내림차순으로 정렬하여 출력해보시오.

- my fellow Koreans, I am grateful to you all. I bow my head in deep appreciation for the 
- choic of you, the people.Today,serving as President in the 19th presidential term of the Republic 
- of Korea, I take the first step toward a new Korea, My shoulders are now burdened with heavy 
- mandates entrusted to me by the poeple,and my heart is burning with enthusiasm to creat
- the kind of country that we have never been able to live in before.My head is not filled with
- blueprints for ushering in a new world characterized by unity coexistence

```python
contents = """
my fellow Koreans, I am grateful to you all. I bow my head in deep appreciation for the
choic of you, the people.Today,serving as President in the 19th presidential term of the Republic
of Korea, I take the first step toward a new Korea, My shoulders are now burdened with heavy
mandates entrusted to me by the poeple,and my heart is burning with enthusiasm to creat
the kind of country that we have never been able to live in before.My head is not filled with
blueprints for ushering in a new world characterized by unity coexistence

"""



word_counts = {}

words = contents.split()
print(words)

for word in words:
    if word.lower() in word_counts :
        word_counts[word.lower()] += 1
    else:
        word_counts[word.lower()] = 1

        
#for (k,v) in word_counts.items():
#    print(k,v)

pairs = list(word_counts.items())
print(pairs)

items = []
for(x,y) in pairs:
    items.append([y,x])
    
print(items)    
items.sort(reverse = True)
print(items)

print()
for i in range(0,len(items)):
    print(items[i][1] + ' : ' + str(items[i][0]))
```

### 4.(집합 변수) 다음 프로그램의 결과를 예측해보시오

```python
a = {1,1,2,2}
print(a) #데이터는 한번으로만 간주

a = (1,2,2,3,3,4);b = set(a) ; print(b)
a = [1,2,2,3,3,4];b = set(a) ; print(b)
s = 'hong gil dong'; b=set(s);print(b)
s = {100,200,300,300,400} ; print(s)
s.add(500)
print(s)
s.discard(100)
print(s)
s.clear()
print(s)
s1 = {1,2,3} ; s2 = {3,4,5,6}
print(s1 & s2) # & 교집합 연산자
print(s1 | s2) # | 합집합 연산자
print(s1 - s2) # - 차집합 연산자
print(s1.intersection(s2))
print(s1.union(s2))
print(s1.difference(s2))
print(s1.issubset(s2))
```
