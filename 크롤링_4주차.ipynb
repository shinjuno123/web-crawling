## 1. urllib 사용해서 크롤링 해보기
- urllib는 웹상의 문서나 화일을 다 가져올 수 있다. 
- 많이 사용하는 패키지와 함수 
 - import urllib.request (주로 urlopen()함수와 urlretrieve() 함수를 사용)
 - read(), geturl(), status, getcode(). getheaders(), info, decode() 
 - from urllib.parse import quote_plus 
- 파이썬 화일처리 관련 함수
 - open(), read(), readline(), readlines(), write(), writelines(), close()
- urlretrieve()와 urlopen() 함수를 이용하여 이미지를 저장하는 프로그램 만들어보기
 - 네이버 검색에서 지구 이미지를 검색한 후 이미지주소복사를 선택한다.
 - 주소 : http://cafefiles.naver.net/data34/2008/6/9/282/%C1%F6%B1%B8_aazz95.jpg
 - c:\temp 폴더를 만들어 놓은 후 실습한다. 
- 다음 홈페이지의 뉴스에서 IT 분야의 헤드 타이틀을 urlopen()과 select() 함수를 이용해서 크롤링해보기 
 - 주소 : https://news.daum.net/digital
 - copy selector : #cSub > div > div.section_cate.section_headline > div.item_mainnews > div > strong > a 

```python
from urllib.request import urlopen
from urllib.request import urlretrieve
from bs4 import BeautifulSoup
import requests

img_url = 'http://cafefiles.naver.net/data34/2008/6/9/282/%C1%F6%B1%B8_aazz95.jpg'
urlretrieve(img_url,'c:/temp/earth1.jpg')
print('다운로드 1 완료')
f = urlopen(img_url).read()
sf = open('c:/temp/earth2.jpg', 'wb')
sf.write(f)
sf.close()
print('다운로드 2 완료')
html_url = 'https://www.google.co.kr/'
urlretrieve(html_url, 'c:/temp/g1.html') 
print('다운로드 3 완료')
f = urlopen(html_url).read()
sf = open('c:/temp/g2.html', 'wb')
sf.write(f)
sf.close()
print('다운로드 4 완료')
url = 'http://www.naver.com'
mem = urlopen(url)
print(mem.geturl())
print(mem.status)

#page = urlopen('https://news.daum.net/digital').read() 
page = requests.get('https://news.daum.net/digital')
soup = BeautifulSoup(page.content, 'html.parser')

titles = soup.select('#cSub > div > div.section_cate.section_headline > div.item_mainnews > div > strong > a')
for title in titles:
    print(title.get_text())
```

## 3. 네이버 증권 사이트 크롤링 해보기(크롤링 한 태그안의 태그 크롤링 해보기) 
- 네이버 증권 사이트의 인기 검색 종목 5개 크롤링 해보기
- 주소 : https://finance.naver.com
- 웹페이지 개발자 모드(f12)의 해당 부분에서 오른쪽 마우스 버튼을 누른 후 copy selector를 선택한다. 
- copy selector : #container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr:nth-child(1) > th > a 
 - tr:nth-child(1) 부분에서는 tr 만 선택한다. 
 - 제일 마지막의 th > a 는 사용하지 않는다. 
 - 수정 copy selector : #container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr  
- 태그 안에 포함된 태그를 크롤링 하는 방법 
 - find() 로 밖의 태그를 먼저 추출하고,
 - 추출된 데이타에서 다시 find()나 find_all()로 원하는 부분을 추출한다.
 - 출력을 보기 좋게 하기 위해 split(), strip() 함수를 활용한다. 

```python
import requests 
from bs4 import BeautifulSoup

page = requests.get('https://finance.naver.com')
soup = BeautifulSoup(page.content,'html.parser')

# data = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr > th > a')
# for item in data:
#     print(item.get_text())

# data = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr')
# for item in data:
#     print(item.find('th').get_text())
#     for imsi in item.find_all('td'):
#         print(imsi.get_text())
#     print()    

data = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr')
for item in data:
    print(item.find('th').get_text(), end= ' ')
    for imsi in item.find_all('td'):
        print(imsi.get_text(), end= ' ')
    print()    
```

```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://finance.naver.com/')
soup = BeautifulSoup(page.content,'html.parser')

datas = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr')

for data in datas:
    print(data.find('th').get_text(),end='\t')
    for imsi in data.find_all('td'):
        print(imsi.get_text(),end='\t')
    print()
```
