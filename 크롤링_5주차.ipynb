## 1. 네이버 날씨에서 미세먼지, 초미세먼지, 오존지수 크롤링 해보기
- 네이버 검색창에 날씨라고 입력한다.
- 날씨정보에서 보이는 미세먼지, 초미세먼지, 오존지수정보를 크롤링한다.
 - 크롤링 주소 : https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=%EB%82%A0%EC%94%A8
 - 화면에 보이는 '미세먼지'에서 마우스 오른쪽 버튼을 누른 후 나오는 메뉴에서 '검사'를 선택한다.(개발자 모드)
 - 태그를 분석하면 div 태그 안에 a 태그, dd 태그가 반복해서 3번 들어가 있다.
 - 이런경우 a 태그 관련 내용을 먼저 다 가져와서 저장하고, 다음으로 dd 태그 내용을 다 가져온다.
 - 가져온 태그 내용을 반복문을 이용하여 한번씩 출력하면 된다. 

````python
import requests 
from bs4 import BeautifulSoup

page = requests.get('https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=%EB%82%A0%EC%94%A8')                                          
soup = BeautifulSoup(page.content,'html.parser')

data = soup.find('div', class_='detail_box')
#print(data)

data_a = data.find_all('a')
#print(len(data_a))
data_dd = data.find_all('dd')
#print(data_dd)

for i in range(len(data_a)):
    print(data_a[i].get_text(), data_dd[i].get_text())
````

- 멜론 차트는 1 ~ 50위, 51 ~ 100 두 페이지로 구성되어 있다.
- requests.get() 함수의 응답이 이상하거나 접근이 허용되지 않을 경우
 - 접속하려는 서버에 접속하는 클라이언트의 헤더 정보(User-Agent)를 제공한다.

## 3. 네이버 블로그 여러 페이지 한번에 크롤링 해보기
- 네이버 검색창에  '파이썬' 이라고 입력한 후 검색한 후 블로그 메뉴를 선택한다.(블로그 검색 결과 화면 아래를 보면 1~10 까지 페이지 번호를 선택할 수 있다.)
- 원하는 페이지를 누른 후 개발자 도구를 이용하여 그 페이지서 다음 페이지에 대한  링크 주소를 확인하면서 규칙을 찾는다. 
``` 
 (예) 2페이지 
 https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=
 &post_blogurl_without=&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sm=tab_pge&srchby=all&st=sim&where=post&start=11
 (예) 3페이지
 https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=
 &post_blogurl_without=&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sm=tab_pge&srchby=all&st=sim&where=post&start=21
```  
- 위의 주소를 비교하여 비교하여 다른 부분(검색어와 페이지번호)을 확인한 후 크롤링 할 url을 만든다.
 - 사용자가 입력한 검색어에 해당하는 부분은 키보드에서 입력받을 수 있도록 수정한다. 
 - 페이지의 시작번호도 변수로 처리할 수 있도록 한다. 
- 반목문을 이용하여 url 주소의 페이지를 사용자가 원하는 만큼 변경하면서 다운로드 받는다.

````python
import requests 
from bs4 import BeautifulSoup
from urllib.parse import quote_plus   

page_num = 1
key_word = quote_plus(input('검색어를 입력하세요 ? '))
i = int(input('몇 페이지를 크롤링 할까요 ? ')) 

count =1
last_page = i * 10 - 9
while  page_num < last_page + 1 :
    url ='https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl \
    =&post_blogurl_without=&query={0}&sm=tab_pge&srchby=all&st=sim&where=post&start={1}'.format(key_word, page_num)    
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')

    blogs = soup.find_all('a', class_='sh_blog_title') 
    print('_______________ {} 페이지 결과 ___________ '.format(count))
    for blog in blogs:
        print(blog.get_text())
        print(blog['href'])
    print()
    
    page_num += 10
    count +=1


'''    
https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=
&post_blogurl_without=&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sm=tab_pge&srchby=all&st=sim&where=post&start=1

https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=
&post_blogurl_without=&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sm=tab_pge&srchby=all&st=sim&where=post&start=11    

https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=
&post_blogurl_without=&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sm=tab_pge&srchby=all&st=sim&where=post&start=21    
'''
````

````python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote_plus


key_word = quote_plus(input('검색어 입력 : '))
i =input('몇 페이지 까지 보고 싶나요?') 
page_num = 1
count = 1
last_page = 10*int(i)-9
while(page_num < last_page+1):    
    url = 'https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=&post_blogurl_without=&query={0}&sm=tab_pge&srchby=all&st=sim&where=post&start={1}'.format(key_word,page_num)
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    
    data = soup.find_all('a',class_='sh_blog_title _sp_each_url _sp_each_title')
    for i in data:
        print(i['href'])
        print(i['title'])
    page_num += 10
    count += 1
    


````
