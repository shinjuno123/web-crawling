 ## 3.네이버 증권 사이트 크롤링 해보기(크롤링 한 태그안의 크롤링 해보기

```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://finance.naver.com/')
soup = BeautifulSoup(page.content,'html.parser')

#data = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr > th > a')
#for item in data:
#    print(item.get_text())

data = soup.select('#container > div.aside > div.group_aside > div.aside_area.aside_popular > table > tbody > tr')
for item in data:
    print('{:<10s}'.format(item.find('th').get_text()),end='')
    for imsi in item.find_all('td'):
        print('{:<10s}'.format(imsi.get_text()),end='')
    print()
```
